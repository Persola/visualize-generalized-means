```python
from math import floor
from scipy.stats import hmean
import seaborn as sns
from matplotlib import rcParams
```

```python
def greyscale_hex_codes(how_many):
    '''Returns linear gradient of hex codes'''

    _max = (16 * 16) - 1
    step = _max / how_many
    decimal_vals = (floor(step * i) for i in range(how_many))
    hex_vals = (hex(dec)[2:] for dec in decimal_vals)
    padded_hex_vals = ('0' + val if len(val) == 1 else val for val in hex_vals)
    return [
        '#' + (val * 3)
        for val
        in padded_hex_vals
    ]
```

```python
def map_funk(funk, save_file=False, _range=range(1, 100)):
    '''Plots the result of a function taking two numerical arguments'''

    # x and y are the two arguments, the color is the result
    x_vals = list(reversed(_range)) # reverse because (0, 0) would be in the upper left by default
    y_vals = list(_range)

    result_grid = []
    for x in x_vals:
        result_row = []
        for y in y_vals:
            result_row.append(funk(x, y))
        result_grid.append(result_row)
    fig = sns.heatmap(
        result_grid,
        cmap=list(greyscale_hex_codes(24)),
        cbar=False
    )
    fig.set(yticks=[], xticks=[])
    rcParams['figure.figsize'] = (10, 10)
    if save_file:
        fig.savefig(save_file)
```

```python
def arithmetic_mean(a, b):
    return (a + b) / 2

map_funk(arithmetic_mean)
```

```python
def geometric_mean(a, b):
    return (a * b)**2

map_funk(geometric_mean)
```

```python
def harmonic_mean(a, b):
    return hmean([a, b])

map_funk(harmonic_mean)
```

```python
def create_p_mean(p):
    def p_mean(a, b):
        return ((a**p + b**p)/2)**(1/p)
    return p_mean
```

```python
# quadratic mean (akin to root mean square/standard deviation):

map_funk(create_p_mean(2))
```
